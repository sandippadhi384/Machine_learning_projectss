```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(handle_unknown='ignore',sparse_output=False).set_output(transform="pandas")
transform=ohe.fit_transform(x[["status_type"]])
transform
```

```python
x=pd.concat([transform,x],axis=1).drop(columns=["status_type"])
x
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x.iloc[:,4:])
x.iloc[:,4:]=scaler.transform(x.iloc[:,4:])
x
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init="k-means++",random_state=42)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title("The Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,init="k-means++",random_state=42)
y_kmeans=kmeans.fit_predict(x)
```

```python
y_kmeans[1:200]
```

```python
x["y_kmeans"]=y_kmeans
```

```python
x
```

```python
one_x=x.loc[x["y_kmeans"]==0,"num_loves"]
one_y=x.loc[x["y_kmeans"]==0,"num_wows"]
two_x=x.loc[x["y_kmeans"]==1,"num_loves"]
two_y=x.loc[x["y_kmeans"]==1,"num_wows"]
three_x=x.loc[x["y_kmeans"]==2,"num_loves"]
three_y=x.loc[x["y_kmeans"]==2,"num_wows"]
four_x=x.loc[x["y_kmeans"]==3,"num_loves"]
four_y=x.loc[x["y_kmeans"]==3,"num_wows"]
five_x=x.loc[x["y_kmeans"]==4,"num_loves"]
five_y=x.loc[x["y_kmeans"]==4,"num_wows"]
plt.scatter(one_x,one_y,color="red",label="1st cluster")
plt.scatter(two_x,two_y,color="blue",label="2nd cluster")
plt.scatter(three_x,three_y,color="green",label="3rd cluster")
plt.scatter(four_x,four_y,color="yellow",label="4th cluster")
plt.scatter(five_x,five_y,color="black",label="5th cluster")
plt.xlabel("num_loves")
plt.ylabel("num_wows")
plt.legend()
plt.show()
```
